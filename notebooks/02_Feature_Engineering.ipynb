# Feature Engineering
---

## 0. Constants

```python
EDA_DATA_PATH = "../data/processed/eda_data.pkl"
PRICE_DATA = "../data/raw/price_data.csv"
FE_DATA_EXPORT_PATH = "../data/processed/fe_data.pkl"
```

---
## 1. Imports

```python
import pandas as pd
import pickle

from sklearn.preprocessing import OneHotEncoder
```

```python
eda_df = pd.read_pickle(EDA_DATA_PATH)
eda_df.head()
```

```python
date_cols = ["date_activ", "date_end", "date_modif_prod", "date_renewal"]
for col in date_cols:
    eda_df[col] = pd.to_datetime(eda_df[col], format="%Y-%m-%d")

eda_df.head()
```

```python
price_df = pd.read_csv(PRICE_DATA)
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

---
## 2. Feature Engineering

```python
#Create Year columns out of date columns
eda_df['year_modif_prod'] = eda_df['date_modif_prod'].dt.year
eda_df['year_renewal'] = eda_df['date_renewal'].dt.year
```

```python
#Create 
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()
monthly_price_by_id.head()
```

```python
#Get January & December prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

#Change columns names
jan_prices.rename(columns={'price_off_peak_var': 'jan_1', 'price_off_peak_fix': 'jan_2'}, inplace=True)
dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}, inplace=True)
```

```python
#Create January-December difference dataframe
diff = pd.DataFrame()
diff["id"] = jan_prices["id"]
diff["offpeak_diff_dec_january_energy"] = dec_prices['dec_1'] - jan_prices['jan_1']
diff["offpeak_diff_dec_january_power"] = dec_prices['dec_2'] - jan_prices['jan_2']
diff.head()
```

```python
eda_df = eda_df.merge(diff, on='id', how='left')
eda_df.head(3)
```

### Drop unnecessary columns

```python
eda_df.info()
```

```python
#drop unnecessary columns
cols_to_drop = ["id", 'date_activ','date_end','date_modif_prod','date_renewal']
```

```python
train_df = eda_df.copy()
train_df = train_df.drop(columns=cols_to_drop)
train_df.head()
```

### Transform columns

```python
#one-hot encode categorical columns
cols_to_transform = ["channel_sales", "has_gas", "origin_up"]
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)

ohe_df = pd.DataFrame(ohe.fit_transform(train_df[cols_to_transform])) 
ohe_df.columns = ohe.get_feature_names_out()
ohe_df.index = train_df.index
ohe_df.head()
```

```python
train_df = train_df.drop(cols_to_transform, axis=1)
train_df = pd.concat([train_df, ohe_df], axis=1)
train_df.head()
```

## 3. Save Data

```python
with open(FE_DATA_EXPORT_PATH, "wb") as file:
    pickle.dump(train_df, file)
```
